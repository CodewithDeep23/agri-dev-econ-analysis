# Soil conservation score

# 🧮 Soil Conservation Score – Explanation

The **Soil Conservation Score** is a heuristic-based scoring system designed to assess the potential for soil conservation on agricultural plots based on various land and soil characteristics. The score ranges between **0.0 to 1.0**, where:

- **Score ≥ 0.6** → Soil conservation is considered **satisfactory** (`"Yes"`).
- **Score < 0.6** → Soil conservation is considered **inadequate** (`"No"`).

---

## 📊 Scoring Factors

| **Feature**           | **Weight**     | **Description** |
|-----------------------|----------------|------------------|
| **Soil Type**         | ±0.2           | High-risk soils (e.g., *Sandy*, *Shallow*) decrease the score, while stable soils (e.g., *Deep black*) increase it. |
| **Soil Depth**        | ±0.2           | Deep soils are less prone to erosion and contribute positively; shallow soils reduce the score. |
| **Soil Fertility**    | ±0.1           | Higher fertility supports vegetation growth and reduces erosion risk. |
| **Slope**             | -0.3 to +0.1   | Steep slopes accelerate erosion (`High slope = -0.3`), while leveled land provides stability. |
| **Soil Binding**      | +0.2           | Indicates the presence of good root binding or structural stability in the soil. |
| **Bunding Presence**  | +0.1           | Bunds help in water retention and preventing runoff, hence improving conservation. |
| **Irrigation Area**   | +0.1           | Higher irrigated area typically implies more vegetative cover, reducing erosion risk. |

---

## 🧠 Score Calculation Logic

- A **baseline score of 0.5** is used.
- Each factor adjusts this baseline based on its contribution.
- Final score is **clipped between 0.0 and 1.0** to ensure it stays valid.
- If final score is **≥ 0.6**, the system marks the field as having **good soil conservation**.

---

## Estimating Soil Loss using USLE

We'll estimate **soil loss** using the **Universal Soil Loss Equation (USLE)** for each plot.

$$
A = R \cdot K \cdot LS \cdot C \cdot P
$$

Where:
- **A**: Average annual soil loss (tons/acre/year)
- **R**: Rainfall erosivity factor (based on climate data)
- **K**: Soil erodibility factor (depends on soil type)
- **LS**: Slope length & steepness factor (based on terrain/slope)
- **C**: Cover & crop management factor (vegetative cover)
- **P**: Support practice factor (bunding, terracing, contouring, etc.)

---

## 📘 Step-by-Step K-Factor Estimation (Qualitative Approach)

### 1. **Soil Texture** (from `soil_type` and `soil_type_ot`)
We classify soils into approximate **texture groups** to estimate the base K-value.

| **Soil Type**                                | **Texture Group**   | **USDA Typical K-Factor Range (approx)** |
|---------------------------------------------|---------------------|------------------------------------------|
| Red soils, Sandy soils, Murrum              | Sandy/Loamy         | 0.05 – 0.20                              |
| Medium black, Deep black                    | Clay/Clay loam      | 0.20 – 0.35                              |
| Problematic, Shallow, Saline/Alkaline       | Mixed/Degraded      | 0.15 – 0.30                              |
| Others, Custom types (from `soil_type_ot`)  | Case-by-case        | Variable                                 |

**🔍 Example Estimates:**
- `Red soils` → Sandy loam → ~**0.17**
- `Deep black` → Clay → ~**0.28**
- `Saline` → Degraded clay loam → ~**0.20**

➡️ A default **base K-value** is assigned based on this classification.

---

### 2. **Soil Depth**
Shallow soils tend to increase runoff and decrease infiltration.

- **< 30 cm** → **Increase** base K by **+0.02**
- **> 100 cm** → **Decrease** base K by **–0.02**

---

### 3. **Soil Fertility** (`soil_fert`)
Used as a proxy for **organic matter (OM)**:

- **Fertility < 30** → **Increase** K by **+0.02**
- **Fertility > 100** → **Decrease** K by **–0.02**

---

### 4. **Soil Degradation** (`soil_degr`)
Used to infer soil structure and overall condition:

- If `soil_degr` contains **"Soil erosion"** or **"Nutrient depletion"** → **Increase** K by **+0.02**
- If `soil_degr` is **"No problem"** → No adjustment


```python
def estimate_k_factor(row):
    # 1. Base K values based on soil type
    soil_type_map = {
        'Red soils': 0.17,
        'Sandy soils': 0.15,
        'Murrum soils': 0.16,
        'Medium black': 0.25,
        'Deep black': 0.28,
        'Shallow': 0.22,
        'Problematic soils': 0.23,
        'Saline/alkaline': 0.20,
        'Others': 0.20,
        0: 0.20,
        None: 0.20
    }
    
    base_k = soil_type_map.get(row['soil_type'], 0.20)

    # 2. Adjust based on soil depth
    depth = row.get('soil_depth', None)
    if pd.notnull(depth):
        if depth < 30:
            base_k += 0.02
        elif depth > 100:
            base_k -= 0.02

    # 3. Adjust based on soil fertility (as a proxy for OM)
    fert = row.get('soil_fert', None)
    if pd.notnull(fert):
        if fert < 30:
            base_k += 0.02
        elif fert > 100:
            base_k -= 0.02

    # 4. Adjust for soil degradation
    degr = row.get('soil_degr', '').lower() if pd.notnull(row.get('soil_degr')) else ''
    if 'erosion' in degr or 'nutrient' in degr:
        base_k += 0.02

    # Ensure K remains within logical range
    return round(min(max(base_k, 0.05), 0.45), 3)
```

```python
agri_df['K_factor'] = agri_df.apply(estimate_k_factor, axis=1)
```

## 🌧 Rainfall Erosivity Factor (R-Factor) Calculation

### 📌 Objective:
To estimate the *Rainfall Erosivity Factor (R)* using available rainfall data (monthly totals, rainy days, and monsoon season totals) from the dataset, based on a custom empirical model derived from literature and regional rainfall characteristics.

---

### 🧪 Formula Used:
$$
R = 50 + 0.3P + 15 \cdot \text{MRR} + 2 \cdot \text{ARI}
$$

Where:
- \( P \) = Total Annual Rainfall (mm) = \( \sum_{i=1}^{12} \text{rain}_i \)
- \( \text{MRR} \) = Monsoon Rainfall Ratio = \( \frac{\text{JJAS\_rain}}{P} \)
- \( \text{ARI} \) = Average Rainfall Intensity = \( \frac{P}{\sum_{i=1}^{12} \text{rd}_i} \)

---

### 📋 Explanation of Components:
| Component | Description | Role in Erosion |
|-----------|-------------|------------------|
| *P* | Total rainfall | Overall erosive volume |
| *MRR* | Proportion of rainfall in monsoon (JJAS) | Captures seasonal storm concentration |
| *ARI* | Rainfall per rainy day | Proxy for intensity (splash energy) |
| *50 (base)* | Constant intercept | Accounts for baseline erosion potential |

---

### 📊 Erosivity Risk Categories:
To interpret R values, we use the following general classification (may vary slightly by region):

| R-Factor Range | Erosion Risk |
|----------------|--------------|
| < 250          | Very Low     |
| 250 – 500      | Low          |
| 500 – 750      | Moderate     |
| 750 – 1000     | High         |
| > 1000         | Very High    |

---

### 📚 Reference Ranges from Literature:
- Semi-Arid India (Rajasthan): 100 – 400
- Central India: 300 – 700
- NE India & Western Ghats: 1000 – 3000
- Lal’s (1976) Equation: \( R = 79 + 0.363P \)

---

### ✅ Conclusion:
The custom formula balances total rainfall, intensity, and monsoon concentration, allowing realistic estimation of R where EI30 values are unavailable. The results can be used for erosion risk assessment or integrated into RUSLE for soil loss modeling.

## 🌱 Converting Slope Categories to LS Factor (Slope Length & Steepness)

In the Universal Soil Loss Equation (USLE), the *LS factor* accounts for the impact of slope length and steepness on erosion. Since our dataset contains slope categories instead of actual gradient values, we assign approximate LS values based on standard erosion modeling practices.

| Slope Category   | Assumed LS Value | Notes |
|------------------|------------------|-------|
| Levelled       | 0.1              | Flat or very gentle land |
| Slight Slope   | 0.3              | Gentle rolling terrain |
| Medium Slope   | 0.5              | Moderate slopes |
| High Slope     | 1.0              | Steep terrain |
| NaN            | 0.3              | Assumed as Slight Slope where data is missing |

This mapping provides a reasonable estimate for LS where precise slope measurements are unavailable.

```python
# Define the slope to LS factor mapping
ls_mapping = {
    'Levelled': 0.1,
    'Slight Slope': 0.3,
    'Medium Slope': 0.5,
    'High Slope': 1.0
}

# Apply the mapping to the slope column to create a new LS factor column
agri_df['LS_factor'] = agri_df['slope'].map(ls_mapping)
```

## P Factor (Support Practice Factor) Estimation

The P factor reflects the effectiveness of soil conservation practices. Based on the dataset, the following columns are used:

- bunding: Indicates whether bunding is practiced ('yes' or 'no').
- bund_type: Numeric code indicating the type of bund.
- bund_type_ot: Descriptions of bunds in text form, for custom or additional detail.

### Mapping Logic

1. If bunding = 'no' or missing → P = 1.0
2. If bunding = 'yes', P is assigned based on bund_type or bund_type_ot:

#### From bund_type (numeric codes):
| bund_type | Description           | P value |
|-----------|------------------------|---------|
| 1         | Contour Bunding        | 0.55    |
| 2         | Graded Bunding         | 0.60    |
| 3         | Earthen Bunds          | 0.70    |
| 4         | Stone Bunds            | 0.50    |

#### From bund_type_ot (text values):
- "CONTOUR" → 0.55
- Contains "STONE" → 0.50
- Contains "FIELD", "PROP", "SOIL CONS" → 0.60
- Contains "EROSION" or "SILT" → 0.60
- "SMALL", "LOOSE" → 0.70
- Else → 0.65 (default for unknown)

3. If both bund_type and bund_type_ot are missing, but bunding is yes → Use default bund P = 0.65

```python
import numpy as np

def assign_p(row):
    # If bunding is 'no' or missing
    if pd.isna(row['bunding']) or row['bunding'].lower() == 'no':
        return 1.0

    # If numeric bund_type is available
    bt = row['bund_type']
    if not pd.isna(bt):
        if bt == 1:
            return 0.55
        elif bt == 2:
            return 0.60
        elif bt == 3:
            return 0.70
        elif bt == 4:
            return 0.50

    # Text-based fallback on bund_type_ot
    bto = str(row['bund_type_ot']).upper() if not pd.isna(row['bund_type_ot']) else ""
    if "CONTOUR" in bto:
        return 0.55
    elif "STONE" in bto:
        return 0.50
    elif any(x in bto for x in ["FIELD", "PROP", "SOIL CONS", "PROPERTY", "SOIL EROSION", "SILT"]):
        return 0.60
    elif any(x in bto for x in ["SMALL", "LOOSE"]):
        return 0.70

    # Default fallback for unknown descriptions
    return 0.65
```

```python
agri_df['P_factor'] = agri_df.apply(assign_p, axis=1)
```

## 🌾 C-Factor Calculation: Cover and Management Factor

The **C-Factor (Cover and Management Factor)** is a crucial component in the Universal Soil Loss Equation (USLE) and Revised USLE (RUSLE). It quantifies the effect of cropping and management practices on soil erosion rates. Lower values indicate greater protection of the soil surface.

### 🔍 Approach:

In this study, we estimated the C-factor by **assigning standard values** to different crop types listed in the dataset. The values were determined based on:

- Published erosion studies
- RUSLE guidelines
- Literature on vegetative cover and erosion risk

Each crop was mapped to a corresponding C-factor range based on its potential for soil protection.

### 📊 Crop-to-C-Factor Mapping:

| Crop Type Category            | Example Crops (from dataset)                                 | Assigned C-Factor | Erosion Risk  |
|------------------------------|---------------------------------------------------------------|-------------------|---------------|
| Dense Cover / Perennial      | Lucern Grass, Barseem, Orchard, Fodder Grass                  | 0.01 – 0.05       | Very Low      |
| Legumes & Cereals            | Chickpea, Pigeonpea, Sorghum, Maize, Blackgram, Greengram     | 0.2 – 0.3         | Moderate      |
| Vegetables & Spices          | Tomato, Brinjal, Onion, Chillies, Coriander, Cucumber         | 0.3 – 0.5         | Moderate–High |
| Oilseeds & Commercial Crops  | Cotton, Castor, Groundnut, Sunflower, Tobacco, Sugarcane      | 0.4 – 0.6         | High          |
| Fallow / Low Cover Areas     | Fallow, Current Fallow, Sparse cover crops                    | 0.6 – 0.8         | Very High     |

> **Note:** For crops not listed in common RUSLE references, an estimated C-value was assigned based on similar crop characteristics.



```python
low_crops = [
    'Paddy', 'Fodder', 'Grass', 'Lucern Grass', 'Barseem',
    'Egyptian Clover', 'Sesbania Bispinosa', 'Dhaincha', 'Tree',
    'Orchard (Bagicha)', 'Jungle Tree'
]
c_factor_low = 0.05

medium_crops = [
    'Cotton', 'Pigeonpea', 'Castor', 'Sorghum', 'Groundnut', 'Maize', 'Pearl Millet',
    'Horsegram', 'Chickpea', 'Blackgram', 'Greengram', 'Wheat', 'Mustard',
    'Lentil', 'Barley', 'Oats', 'Rapeseed', 'Lathyrus', 'Peas', 'Cowpea',
    'Soybean', 'Moong', 'Minor Pulse', 'Tiwara', 'Phaseolu Trilobus'
]
c_factor_medium = 0.3

high_crops = [
    'Chillies', 'Vegetables', 'Tomato', 'Brinjal', "Lady's Finger", 'Onion', 'Coriander',
    'Cucumber', 'Cluster Bean', 'Gherkin', 'Bitter Gourd', 'Ridge Gourd', 'Potato',
    'Bottle Gourd', 'Radish', 'Cauliflower', 'Garlic', 'Beans', 'Green Chilli',
    'Cabbage', 'Carrot', 'Turmeric', 'Cumin', 'Ginger', 'Guar gum', 'Ivy Gourd',
    'Flower', 'Colocasia', 'Pumpkin', 'Capsicum', 'Papaya', 'Leafy Vegetable',
    'Amaranthus', 'Green Pea', 'Gardenpea', 'Fenugreek', 'Chrysanthemum',
    'China Aster', 'Marigold', 'Red Green Leaf', 'Green Leaves'
]
c_factor_high = 0.5

very_high_crops = [
    'Current Fallow', 'Fallow', 'Banana', 'Mango', 'Orange', 'Guava', 'Pomegranate',
    'Citrus', 'Coconut', 'Teak', 'Betel Vine', 'Acarus Calamus', 'Lotus Seed',
    'Custard Apple', 'Hulga', 'Gourd', 'Sapota', 'Sarguja', 'Jasmine',
    'Billion Dollar Grass (Sawan)', 'Avare', 'Watermelon', 'Mahua', 'Kagzi Lime',
    'Bokala', 'Rabda', 'Arecanut', 'Safflower', 'Agashi', 'Tobacco', 'Subabul'
]
c_factor_very_high = 0.85

# Default c_factor
c_factor_default = 0.6
```

```python
def assign_c_factor(crop):
    crop = str(crop).strip()
    if crop in low_crops:
        return c_factor_low
    elif crop in medium_crops:
        return c_factor_medium
    elif crop in high_crops:
        return c_factor_high
    elif crop in very_high_crops:
        return c_factor_very_high
    else:
        return c_factor_default

agri_df['C_factor'] = agri_df['crop'].apply(assign_c_factor)
```

```python
agri_df['A_factor'] = (
    agri_df['R_factor'] *
    agri_df['K_factor'] *
    agri_df['LS_factor'] *
    agri_df['C_factor'] *
    agri_df['P_factor']
)
```
